```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
from bs4 import BeautifulSoup, SoupStrainer
import sys, string, requests, datetime, time
%matplotlib inline
```

```python
players_info = pd.read_csv('nba_players_info.csv')
```

```python
players_info.head()
```

```python
players_info.describe()
```

```python
left = []
right = []
height = []
weight = []
ppg = []
trb = []
ast = []

start = players_info.active_from.min()
end = players_info.active_from.max()

for year in range(start,end):
    active = (players_info.active_from.values <= year) & (players_info.active_to >= year) # only include active players
    left.append(1.0*np.sum(players_info[active].shooting_hand == 'Left')) # number of left handed shooters
    right.append(1.0*np.sum(players_info[active].shooting_hand == 'Right')) # number of right handed shooters
    height.append(np.mean(players_info[active].height)) # average height in inches
    weight.append(np.mean(players_info[active].weight)) # average weight in pounds
    ppg.append(np.mean(players_info[active].ppg))
    trb.append(np.mean(players_info[active].trb))
    ast.append(np.mean(players_info[active].ast))
    
# create a DataFrame with a summary
years = pd.DataFrame({'season': np.arange(start,end), 
                      'right': right,
                      'left': left,
                      'height': height,
                      'weight': weight,
                      'ppg': ppg,
                      'trb': trb,
                      'ast': ast}).set_index('season')

# preview
years.head()
    
```
