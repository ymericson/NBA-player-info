# 911 Calls Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Tables/911.csv')
df.head(10)
```

```python
df['Reason'] = df['title'].apply(lambda x: x.split(":")[0])
df.head(10)
```

```python
df['Reason'].value_counts()
```

```python
sns.countplot(x='Reason', data=df)
plt.show()
```

```python
df['timeStamp'] = pd.to_datetime(df['timeStamp'])
df['Hour'] = df['timeStamp'].apply(lambda x: x.hour)
df['Day of Week'] = df['timeStamp'].apply(lambda x: x.dayofweek)
df['Month'] = df['timeStamp'].apply(lambda x: x.month)
df.head()
```

```python
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
df['Day of Week'] = df['Day of Week'].map(dmap)
df.head()
```

```python
sns.countplot(x='Day of Week', data = df, hue = 'Reason', palette='viridis')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
sns.countplot(x='Month', data = df, hue = 'Reason', palette='viridis')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
byMonth = df.groupby('Month').count()
byMonth.head(10)
```

```python
# count of calls per month
byMonth['lat'].plot()
```

```python
sns.lmplot(x='Month', y='twp', data=byMonth.reset_index())
```

```python
df['Date'] = df['timeStamp'].apply(lambda x: x.date())
df.head()
```

```python
byDate = df.groupby('Date').count()
byDate.head()
```

```python
df.groupby('Date').count()['lat'].plot(figsize=(8,4))
plt.title('All Calls')
plt.tight_layout()
```

```python
df[df['Reason']=='Traffic'].groupby('Date').count()['lat'].plot(figsize=(8,4))
plt.title('Traffic')
plt.tight_layout()
```

```python
df[df['Reason']=='Fire'].groupby('Date').count()['lat'].plot(figsize=(8,4))
plt.title('Fire')
plt.tight_layout()
```

```python
df[df['Reason']=='EMS'].groupby('Date').count()['lat'].plot(figsize=(8,4))
plt.title('EMS')
plt.tight_layout()
```

```python
dayHour = df.groupby(by=['Day of Week', 'Hour']).count()['Reason'].unstack()
dayHour
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(dayHour, cmap='coolwarm')
plt.tight_layout()
```

```python
sns.clustermap(dayHour, cmap='coolwarm')
```

```python
dayMonth = df.groupby(by=['Day of Week', 'Month']).count()['lat'].unstack()
dayMonth
```

```python
plt.figure(figsize=(9,7))
sns.heatmap(dayMonth, cmap='coolwarm')
```

```python
sns.clustermap(dayMonth, cmap='coolwarm')
```
